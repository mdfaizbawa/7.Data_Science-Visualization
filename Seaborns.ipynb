```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Placement.csv')
dataset
```

```python
#Number of rows and columns
print('no. of row',dataset.shape[0])
print('no. of column',dataset.shape[1])
```

#Univariate Analysis

```python
#Distplot:- 
```

```python
sb.distplot(dataset["ssc_p"],hist=True, kde=True, kde_kws={'color':'Red'})
```

```python
#To find the histogram= frequency(repeation of data)
```

#Bivariate Analysis : Distribution Plot

```python
#Jointplot
```

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset)
plt.show()
```

```python
#To difine the scatter plot between ssc_p and salary.
#To show the histrogram X-axis and Y-axis
```

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='hex')
plt.show()
```

```python
#hex= hexigon or heat map
#if more drak shows more repeation
#if less dark shows less repeation
```

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='kde')
plt.show()
#kde= kernel density estimation
```

```python
#PairPlot
```

```python
sb.pairplot(dataset,hue='gender',diag_kind="kde",kind='scatter',palette='husl')
plt.show()
```

```python
#For Categorical plot: (Categorical vs Numerical)
```

```python
from matplotlib import pyplot as plt
df = sb.load_dataset('exercise')
sb.catplot(x = "time", y = "pulse", hue = "kind", kind = 'violin', col = "diet", data = df);
plt.show()
```

```python
df = sb.load_dataset('tips')
sb.regplot(x = "total_bill", y = "tip", data = df)
sb.lmplot(x = "total_bill", y = "tip", data = df)
plt.show()
```

```python
df = sb.load_dataset('anscombe')
sb.lmplot(x = "x", y = "y", data = df.query("dataset == 'II'"))
plt.show()
```

```python
df = sb.load_dataset('anscombe')
sb.lmplot(x = "x", y = "y", data = df.query("dataset == 'II'"),order = 2)
plt.show()
```

```python
df = sb.load_dataset('tips')
g = sb.FacetGrid(df, col = "time")
g.map(plt.hist, "tip")
plt.show()
```

```python
df = sb.load_dataset('iris')
g = sb.PairGrid(df)
g.map(plt.scatter);
plt.show()
```
